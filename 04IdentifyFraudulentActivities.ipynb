Task 1: For each user, determine her country based on the numeric IP address.

```python
import pandasql as ps

sqlcode = '''
select A.*,B.country
from data A
left join dic B on a.ip_address
between b.lower_bound_ip_address and b.upper_bound_ip_address
'''

new_data = ps.sqldf(sqlcode,locals())
```

```python
new_data['country']=new_data.country.fillna('Missing')
```

```python
new_data
```

Machine learning model - Logistic regression

```python
from sklearn.linear_model import LogisticRegressionCV
from sklearn.model_selection import train_test_split

x=new_data[['purchase_value','source','browser','sex','age','Min_diff','country']]
x=pd.get_dummies(x)
y=new_data['class'].astype(str)

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
lr=LogisticRegressionCV(scoring='roc_auc',max_iter=1000,class_weight='balanced')
lr.fit(x_train,y_train)
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix

ytest_predict = lr.predict(x_test)
print(classification_report(y_true=y_test,y_pred=ytest_predict))
```

```python
importance=pd.DataFrame({'var name':x_train.columns,'coef':lr.coef_[0]})
importance['importance']=abs(importance.coef)
importance.sort_values(by=['importance'], ascending=False).head(20)
```

Change the cutoff(default at .5)

reference code: https://stackoverflow.com/questions/28716241/controlling-the-threshold-in-logistic-regression-in-scikit-learn

```python
from sklearn.metrics import accuracy_score

pred_proba_df = pd.DataFrame(lr.predict_proba(x_test))
threshold_list = [0.05,0.1,0.15,0.2,0.25,0.3,0.35,0.4,0.45,0.5,0.55,0.6,0.65,.7,.75,.8,.85,.9,.95,.99]
for i in threshold_list:
    print ('\n******** For i = {} ******'.format(i))
    Y_test_pred = pred_proba_df.applymap(lambda x: 1 if x>i else 0)
    test_accuracy = accuracy_score(y_test.astype(int).to_numpy().reshape(y_test.to_numpy().size,1),
                                           Y_test_pred.iloc[:,1].to_numpy().reshape(Y_test_pred.iloc[:,1].to_numpy().size,1))
    print('Our testing accuracy is {}'.format(test_accuracy))

    print(confusion_matrix(y_test.astype(int).to_numpy().reshape(y_test.to_numpy().size,1),
                           Y_test_pred.iloc[:,1].to_numpy().reshape(Y_test_pred.iloc[:,1].to_numpy().size,1)))
```
